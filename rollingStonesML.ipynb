```python
def classifica_musica_ao_vivo(df_rs):
    if df_rs["liveness"]>= 0.8:
        return True
    else:
        return False
```

```python
df_rs["ao_vivo"] = df_rs.apply(classifica_musica_ao_vivo, axis=1)
```

```python
df_rs
```

```python
df_rs.groupby("ao_vivo")["ao_vivo"].count()
```

```python
df_gravado_em_estudio = df_rs[df_rs["ao_vivo"]==False]
df_show_ao_vivo = df_rs[df_rs["ao_vivo"]==True]
```

```python
print("Média das músicas ao vivo: ",df_show_ao_vivo["duracao_em_min"].mean())
```

```python
print("Média das músicas gravadas em estúdio: ",df_gravado_em_estudio["duracao_em_min"].mean())
```

```python
df_studio = df_gravado_em_estudio.groupby("album")["loudness"].sum()
df_ao_vivo = df_show_ao_vivo.groupby("album")["loudness"].sum()
```

```python
fig, axes = plt.subplots(1,2,figsize = (15,4))

sns.histplot(data=df_studio, bins=20, ax=axes[0])
axes[0].set_title("Soma do barulho dos álbuns de estúdio")
axes[0].set_xlabel("Barulho")
axes[0].set_ylabel("Frequência")

sns.histplot(data=df_ao_vivo, bins=20, ax=axes[1])
axes[1].set_title("Soma do barulho dos álbuns ao vivo")
axes[1].set_xlabel("Barulho")
axes[1].set_ylabel("Frequência")

fig.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.kdeplot(data=df_studio, label="Álbuns de estúdio", fill = True)
sns.kdeplot(data=df_ao_vivo, label="Álbuns ao vivo", fill = True)

plt.title("Distribuição do barulho dos álbuns")
plt.xlabel("Barulho")
plt.ylabel("Densidade")
plt.legend()
```

```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro(df_studio)#teste de normalidade
print("Soma dos barulho dos álbuns de estúdio ")
print("Estatística de teste: {:.4f}, valor p: {}".format(stat,p))

if p > 0.05:
    print("Não há evidência suficiente para rejeitar a hipótese de normalidade.")
else:
    print("A hipótese de normalidade é rejeitada.")
```

```python
stat, p = shapiro(df_ao_vivo)#teste de normalidade
print("Soma dos barulho dos álbuns ao vivo ")
print("Estatística de teste: {:.4f}, valor p: {}".format(stat,p))

if p > 0.05:
    print("Não há evidência suficiente para rejeitar a hipótese de normalidade.")
else:
    print("A hipótese de normalidade é rejeitada.")
```

```python
from scipy.stats import mannwhitneyu
```

```python
stat, p = mannwhitneyu(df_studio.sample(len(df_studio)),df_ao_vivo.sample(len(df_ao_vivo)), alternative = "less")

print("Estatística de teste U: ", stat)
print("Valor p: ", p )

alpha = 0.05
if p < alpha:
    print("Diferença estatisticamente significante.")
else:
    print("Não há diferença estatisticamente significante")
```

```python
#Correlação 
media_por_album = df_rs.groupby("album")["valence"].mean().reset_index()

media_por_album = media_por_album.rename(columns={"valence": "media_valence"})
media_por_album["sentimento"] = ["positivo" if v >0.6 else "negativo" for v in media_por_album["media_valence"]]

media_por_album.groupby("sentimento")["sentimento"].count()
```

```python
media_por_album.head()
```

```python
df_resultado_final = pd.merge(df_rs, media_por_album, on = "album")
df_resultado_final.head()
```

```python
matriz_correlacao = df_resultado_final.corr(numeric_only=True)

correlacao_sentimento = matriz_correlacao["media_valence"]

display(correlacao_sentimento)
```

```python
sns.heatmap(correlacao_sentimento.to_frame(), annot=True, cmap= "coolwarm")
plt.show()
```

```python
sns.scatterplot(x = "media_valence", y = "danceability", hue = "sentimento", data = df_resultado_final, palette = "coolwarm")
plt.xlabel("Média de valence")
plt.ylabel("Danceability")
plt.title("Relação entre valência média e a capacidade de dança das músicas")
plt.show()
```

```python
sns.scatterplot(x = "media_valence", y = "liveness", hue = "sentimento", data = df_resultado_final, palette = "coolwarm")
plt.xlabel("Média de valence")
plt.ylabel("Liveness")
plt.title("Relação entre valência média e a probabilidade das músicas serem ao vivo")
plt.show()
```
