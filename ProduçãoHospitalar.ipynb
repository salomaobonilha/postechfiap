```python
def plota_gastos_por_mes(dados,figsize=(10,6)):
  axis= dados.T.plot(figsize=figsize)
  axis.set_ylabel("Gasto em milhões de reais")
  axis.set_xlabel("Mês")
```

```python
plota_gastos_por_mes(sorted_dados[meses_filtrados].head(3))
plt.ylim(0,600)
```

```python
plota_gastos_por_mes(sorted_dados[sorted_dados.columns[45:70]].head(3))
plt.ylim(0,340)

```

```python
plota_gastos_por_mes(sorted_dados[sorted_dados.columns[-12:]].head(3))
plt.ylim(80,640)
```
